```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from collections import Counter

# Load dataset
df = pd.read_csv("milknew.csv")

# Encode Grade first (Low=0, Medium/High=1)
df['Grade_encoded'] = df['Grade'].map({'low': 0, 'medium': 1, 'high': 1})

# Features and target
X = df.drop(columns=['Grade', 'Grade_encoded'])
y = df['Grade_encoded']

print("Before SMOTE:", Counter(y))

# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

print("After SMOTE:", Counter(y_resampled))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(1, 2, figsize=(10,4))

sns.countplot(x=y, ax=ax[0], palette="Set2")
ax[0].set_title("Before SMOTE")
ax[0].set_xlabel("Grade Encoded")

sns.countplot(x=y_resampled, ax=ax[1], palette="Set2")
ax[1].set_title("After SMOTE")
ax[1].set_xlabel("Grade Encoded")

plt.show()
```
