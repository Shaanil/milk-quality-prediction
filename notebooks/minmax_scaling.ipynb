```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("milknew.csv")

scaler = MinMaxScaler()

df['Colour_scaled'] = scaler.fit_transform(df[['Colour']])

print(df[['Colour', 'Colour_scaled']].head())

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
sns.histplot(df['Colour'], kde=True, color="skyblue")
plt.title("Colour - Before Scaling")

plt.subplot(1,2,2)
sns.histplot(df['Colour_scaled'], kde=True, color="orange")
plt.title("Colour - After Min-Max Scaling")

plt.tight_layout()
plt.show()
```
